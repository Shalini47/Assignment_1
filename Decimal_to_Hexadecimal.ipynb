```python
# function to convert  
# decimal to hexadecimal
import numpy as np 
def decToHexa():
    
      
    # char array to store  
    # hexadecimal number 
    random_number = np.random.randint(low=0, high=5000, size=1000)
    hexaDeciNum = [''] * 1000; 
    #Print random values
    print("Decimal number\n")
    for j in range(0,1000):
    	print(random_number[j])
    # counter for hexadecimal  
    # number array
    print("\nhexadecimal number\n") 
    for i in range(0,1000):
    	while(random_number[i] != 0): 
          
        # temporary variable  
        # to store remainder 
        	temp = 0; 
          
        # storing remainder  
        # in temp variable. 
        	temp = random_number[i]  % 16; 
          
        # check if temp < 10 
        	if(temp < 10): 
        		hexaDeciNum[i] = hexaDeciNum[i]+chr(temp + 48); 
            
        	else: 
        		hexaDeciNum[i] = hexaDeciNum[i]+chr(temp + 55); 
            
        	random_number[i] = int(random_number[i] / 16);
          
    # printing hexadecimal number 
    # array in reverse order
    	hexaDeciNum[i] = "".join(reversed(hexaDeciNum[i]))
    	print(hexaDeciNum[i])

 
# Driver Code 
decToHexa();
```
